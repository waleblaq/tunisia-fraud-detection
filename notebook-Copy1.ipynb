```python
ready_train = fulldata[:ntrain]
ready_test = fulldata[ntrain:]
```

```python
X= ready_train.drop(['id', 'target', 'BCT_CODBUR', 'CTR_MATFIS', 'ACT_CODACT', 'FJU_CODFJU'], axis = 1)
y = ready_train['target']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
p = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

rms = sqrt(mean_squared_error(y_test, p))
rms
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators = 550)
rfr.fit(X_train, y_train)
```

```python
pr = rfr.predict(X_test)
```

```python
from sklearn.model_selection import KFold, cross_val_score
n_folds = 5
def rmsle_cv(model):
    kf = KFold(n_folds, shuffle=True, 
               random_state=42).get_n_splits(X.values)
    rmse= np.sqrt(-cross_val_score(model, X.values, y.values.ravel(), 
                  scoring="neg_mean_squared_error", cv = kf))
    return(rmse)

score = rmsle_cv(rfr)
print("Random Forest score: {:.3f} ({:.3f})\n".format(score.mean(), 
       score.std()))
```
